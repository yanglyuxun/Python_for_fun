```python
import pandas as pd
```

```python
data=pd.read_html('list.html')[0]
data.index = data['#']

data
```

```python
data = data.iloc[:,2:]
def str2num(s):
    return float(s[:-2])
data['Acceptance'] = data['Acceptance'].apply(str2num)
data = data.sort_values(by='Acceptance',ascending=False)
data['Done_online'] = False
data
```

```python
from bs4 import BeautifulSoup
with open('list.html') as f:
    soup = BeautifulSoup(f.read(), 'html.parser')
```

```python
tb = soup.table
```

```python
# column names
rows = tb.find_all('tr')
rows[0].find_all('th')
colname = [a.text.strip() for a in rows[0].find_all('th')]
colname[0] = 'Done_online'
colname.pop(1)
data = pd.DataFrame(columns=colname)
data
```
