```python
# column names
rows = tb.find_all('tr')
rows[0].find_all('th')
colname = [a.text.strip() for a in rows[0].find_all('th')]
colname[0] = 'Done_online'
colname.pop(1)
data = pd.DataFrame(columns=colname)
data
```

```python
for row in rows[1:-1]:
    cols = row.find_all('td')
    ind = int(cols[1].text)
    data.loc[ind,'Done_online'] = cols[0]['value'] if 'value' in cols[0].attrs else None
    data.loc[ind,'Title'] = cols[2].attrs['value']
    data.loc[ind,'Lock'] = True if cols[2].span else False
    data.loc[ind,'problem_url'] = cols[2].a['href'] if cols[2].a else None
    data.loc[ind,'artivle_url'] = cols[3].a['href'] if cols[3].a else None
    data.loc[ind,'Acceptance'] = float(cols[4]['value'])
    data.loc[ind,'Difficulty'] = cols[5].text.strip()
data
```

```python
data.to_json('data.json')
```
